```python

import ast
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()                
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
import ast

def convert3(text):
    L = []
    text2 = ast.literal_eval(text)  
    for item in text2[:3]:         
        L.append(item['name'])      
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert)
movies.head()
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new['tags']=new['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
```

```python
def recommend(movie, sort_by_rating=False):
    # Find index of given movie
    index = new[new['title'] == movie].index[0]
    
    # Get all similarity scores for this movie
    distances = sorted(
        list(enumerate(similarity[index])),
        reverse=True,
        key=lambda x: x[1]
    )
    
    # Get top 5 similar movies (skip the movie itself)
    top_movies = [
        (i[0], new.iloc[i[0]].title, new.iloc[i[0]].vote_average)
        for i in distances[1:6]
    ]
    
    if sort_by_rating:
        # Sort by rating in descending order
        top_movies.sort(key=lambda x: x[2], reverse=True)
        print("Top similar movies sorted by rating:")
        for _, title, rating in top_movies:
            print(f"{title} ({rating})")
    else:
        # Keep similarity order, just show rating in parentheses
        for _, title, rating in top_movies:
            print(f"{title} ({rating})")
```

```python
recommend('Batman',True)
```
